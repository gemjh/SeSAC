- 비용이 많이 드는 함수 호출의 결과를 저장하고 동일한 입력이 다시 발생할 때 캐시된 결과를 반환

### st.cache_data
- 데이터를 반환하는 계산을 캐싱할 때 사용
- 직렬화 가능한 데이터 객체(예: str, int, float, DataFrame, dict, list)를 반환하는 함수에 사용



### st.cache_resource
- ML 모델이나 데이터베이스 연결과 같은 전역 리소스를 캐싱할 때 사용

- 예시
```
@st.cache_data
def long_running_function(param1, param2):
    return …
```

## 2. Session State

### Session 이란
- 세션은 앱을 보는 단일 인스턴스
- 브라우저에서 두 개의 다른 탭에서 앱을 보면 각 탭마다 자체 세션이 있음
- Streamlit은 사용자가 앱과 상호 작용하는 동안 이 세션을 유지
- 사용자가 브라우저 페이지를 새로 고치거나 앱의 URL을 다시 로드하면 세션 상태가 재설정되고 새 세션으로 다시 시작

### Session State
- Session State는 딕셔너리와 같은 인터페이스를 제공
- st.session_state를 사용하여 값을 저장하고 검색할 수 있음
- st.session_state["my_key"] 또는 st.session_state.my_key와 같이 사용

- 즉, 피커로 다른 색상을 선택하면 데이터가 각 재실행마다 다시 무작위로 생성되지 않음
